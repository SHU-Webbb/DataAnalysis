# 지하철 유무임 승차 분석

## 1. csv 파일 읽고 데이터 정제

```python
import pandas as pd

subway_raw = pd.read_csv('./data/subwayfree_202403.csv', encoding='utf-8')
subway_raw.head()
```

```python
subway_raw.info()
```

### (1) 유무임 승차 비율이 가장 높은 역
#### 첫번째 방법: 유임승차/무임승차

```python
# 원본 복사
subway_df1 = subway_raw.copy(deep=True)
```

```python
subway_df1 = subway_df1[subway_raw['무임승차'] != 0]
```

```python
subway_df1['유무임승차비율1'] = subway_df1['유임승차'] / subway_df1['무임승차']
```

```python
subway_df1
```

```python
subway_sort = subway_df1.sort_values(by='유무임승차비율1', ascending=False)
```

```python
subway_sort
```

```python
print('유임승차비율이 높은 역: %s, 비율:%f' % (subway_sort.iloc[0]['지하철역'], subway_sort.iloc[0]['유무임승차비율1']))
```

### (2) 유무임 승차 비율이 가장 높은 역
#### 두번째 방법: 무임승차/(유임승차+무임승차)

```python
subway_df2 = subway_raw.copy(deep=True)
```

```python
subway_df2 = subway_df2[subway_raw['무임승차'] != 0]
```

```python
subway_df2['유무임승차비율2'] = subway_df1['유임승차'] / (subway_df1['유임승차'] + subway_df1['무임승차'])
```

```python
subway_df2
```

```python
subway_sort = subway_df2.sort_values(by='유무임승차비율2', ascending=False)
```

```python
print('유임승차비율이 높은 역: %s, 비율:%f' % (subway_sort.iloc[0]['지하철역'], subway_sort.iloc[0]['유무임승차비율2']))
```
