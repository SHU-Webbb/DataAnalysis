```python
import pandas as pd
```

```python
row_data1 = pd.read_csv('row_data1.csv', encoding='cp949')
row_data2 = pd.read_csv('row_data2.csv', encoding='cp949')
```

```python
row_data1.head()
```

```python
row_data1.columns
```

```python
row_data2.head()
```

```python
food_data1 = pd.DataFrame({'이름' : row_data1['식품명'],
                           '탄' : row_data1['탄수화물(g)'],
                           '단' : row_data1['단백질(g)'],
                           '지' : row_data1['지방(g)']})
food_data2 =  pd.DataFrame({'이름' : row_data2['식품명'],
                           '탄' : row_data2['탄수화물(g)'],
                           '단' : row_data2['단백질(g)'],
                           '지' : row_data2['지방(g)']})
```

```python
food_data1.head()
```

```python
food_data2.head()
```

```python
food_data = pd.concat([food_data1, food_data2])
```

```python
food_data
```

```python
food_data = food_data.reset_index()
```

```python
food_data.rename(columns={food_data.columns[0] : '음식ID'},
                          inplace = True)
```

```python
food_data
```

```python
food_data[food_data.columns[0]] = range(1, len(food_data) + 1)
```

```python
food_data
```

```python
food_data.to_csv('food_data1.csv', index=False, encoding='cp949')
```
