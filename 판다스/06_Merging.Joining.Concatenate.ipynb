# Merging, Joining, Concatenate

```python
# 샘플 데이터
import pandas as pd
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']},
                   index = [0,1,2,3])
```

```python
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                    'B':['B4','B5','B6','B7'],
                    'C':['C4','C5','C6','C7'],
                    'D':['D4','D5','D6','D7']},
                   index = [4,5,6,7])
```

```python
df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                    'B':['B8','B9','B10','B11'],
                    'C':['C8','C9','C10','C11'],
                    'D':['D8','D9','D10','D11']},
                   index = [8,9,10,11])
```

## 1. Concatenation

#### concatenation은 데이터 프레임을 이어 붙이는 형식. 차원의 수가 일치해야 한다.
#### pd.concat 명령을 사용하고 이어붙일 데이터 프레임의 리스트를 전달

```python
# 기본적으로 행을 기준으로 이어 붙임.
pd.concat([df1, df2, df3])
```

```python
# 축의 값을 지정
pd.concat([df1, df2, df3], axis=1)
```

## 2. Merge

#### merge는 sql의 join 명령과 유사하게 데이터 프레임을 합침

```python
left = pd.DataFrame({'key':['k0','k1','k2','k3'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']})              
```

```python
right = pd.DataFrame({'key':['k0','k1','k2','k3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})    
```

 ### (1) inner merge

```python
# on - 결합할 기준 컬럼 지정"
pd.merge(left, right, how='inner', on='key')
```

```python
left = pd.DataFrame({'key1':['k0','k0','k1','k2'],
                     'key2':['k0','k1','k0','k1'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']})       
right = pd.DataFrame({'key1':['k0','k1','k1','k3'],
                      'key2':['k0','k0','k0','k0'],
                         'C':['C0','C1','C2','C3'],
                         'D':['D0','D1','D2','D3']})    
```

```python
#how 기본값은 inner임

pd.merge(left, right, on=['key1', 'key2'])
```

### (2) outer merge
#### 한쪽 데이터프레임을 기준으로 merge하는 작업

```python
left
```

```python
right
```

```python
# sql의 fullouter 구조와 유사
pd.merge (left, right, how='outer' , on=['key1', 'key2'])
```

```python
# 기준 데이터프레임을 지정하여 outer merge
pd.merge (left, right, how='right' , on=['key1', 'key2'])
```

```python
pd.merge (left, right, how='left' , on=['key1', 'key2'])
```

## 3.Join

#### join - 서로 다르게 인데스된 데이터프레임에 대해 결합하는 방법

```python
left = pd.DataFrame({'A':['A0','A1','A2'],
                     'B':['B0','B1','B2']},
                   index = ['KO','K1','K2'])       
right = pd.DataFrame({'C':['C0','C2','C3'],
                       'D':['D0','D2','D3']},
                    index = ['K0','K2','K3'])    
```

```python
# 왼쪽 데이터프레임키의 인덱스와 일치하는 값만 join수행
left.join(right)
```

```python
# 왼쪽과 오른쪽 데이터프레임을 모두 가져옴.
left.join(right, how='outer')
```
