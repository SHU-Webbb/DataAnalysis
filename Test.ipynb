```python
import pandas as pd
import numpy as np
seoul_raw = pd.read_csv('./data/Seoul.csv' , encoding='utf-8')
```

```python
seoul_raw
```

```python
seoul_raw.drop(['\t최고기온시각','최저기온시각'], axis = 1, inplace=True)
```

```python
seoul_raw
```

```python
seoul_raw.rename(columns={'평균기온(℃)' : '평균기온',
                          '최고기온(℃)' : '최고기온',
                           '최저기온(℃)' : '최저기온'
                           }, inplace=True)
```

```python
seoul_raw
```

```python
seoul_raw.isnull().sum()
```

```python
seoul_raw.dropna()
```

```python
print(seoul_raw.sort_values(by='최고기온', ascending=False).iloc[0]['일시'],
seoul_raw.sort_values(by='최고기온', ascending=False).iloc[0]['최고기온'])
```

```python
import matplotlib.pyplot as plt
t_high = seoul_raw['최고기온']
plt.figure(figsize=(10,6))
plt.plot(t_high)
plt.show()
```

```python
seoul_0412 = seoul_raw[seoul_raw['일시'].astype(str).str.contains('-04-12')]
seoul_0412
```

```python
high_0412 = seoul_0412['최고기온']
low_0412 = seoul_0412['최저기온']
```

```python
plt.figure(figsize=(10,6))
plt.plot(high_0412, label = 'high')
plt.plot(low_0412, label = 'low')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
card_raw = pd.read_csv('./data/201906.csv' , encoding='utf-8')
card_raw
```

```python
seoul_card = card_raw[card_raw['CSTMR_MEGA_CTY_NM'].astype(str).str.contains('서울특별시')]
print('서울시 거주 고객수:',seoul_card['CSTMR_MEGA_CTY_NM'].count())
not_seoul_card = card_raw[~card_raw['CSTMR_MEGA_CTY_NM'].astype(str).str.contains('서울특별시')]
print('서울시 비거주 고객수:',not_seoul_card['CSTMR_MEGA_CTY_NM'].count())
```

```python
print('서울시 거주 고객의 소비액 :', seoul_card['AMT'].sum())
print('서울시 비거주 고객의 소비액 :', not_seoul_card['AMT'].sum())
```

```python
seoul_1 = seoul_card[seoul_card['CTY_RGN_NM'].astype(str).str.contains('관악구')]
print('관악구 전체 소비액 :',seoul_1['AMT'].sum())
```

```python
print('관악구 편의점 소비액:',seoul_1[seoul_1['TP_BUZ_NO']==4010]['AMT'].sum())
```
